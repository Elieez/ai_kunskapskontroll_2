### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

```python
Figure is the top level container for all elements the overall window that can hold one or more subplots.
Axes is a single plotting area within a figure, each axes has its own X and Y axis, labels and data.
Axis is the indivdiual X or Y axis inside a axes, it handels tick marks, tick labels and the numeric scale along one dimension.
Arists is any object that knows how to draw itself on the figure like Axes, axis, text labels, line markers etc. 
```

### When plotting in Matplotlib, what is the expected input data type?

```python
Matplotlib’s plotting functions expect “array‐like” inputs—typically one‐dimensional sequences.
```

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
x = np.linspace(-4, 4, 10)
y = x**2

# Pyplot
plt.plot(x, y)
plt.title("y = x²")
plt.xlabel("x")
plt.ylabel("y")
plt.show()

# object oriented 
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title("y = x²")
ax.set_xlabel("x")
ax.set_ylabel("y")

plt.show()
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

ax1.scatter(random_data_x, random_data_y, alpha=0.6, edgecolor='k')
ax1.set_title('Scatter Plot (Random Data)')
ax1.set_xlabel('random_data_x')
ax1.set_ylabel('random_data_y')

ax2.bar(names, values, color='orange')
ax2.set_title('Bar Plot (Fruit Counts)')
ax2.set_xlabel('Fruit')
ax2.set_ylabel('Count')

plt.show()
```
